# **1) Análise exploratória de dados**

```python
#importar as bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#Ler o conjunto de dados
data= pd.read_csv('/content/netflix_titles.csv')

#Mostrar os dados
data
```

```python
#Separar os dados por ','
data['listed_in_novo']= data['listed_in'].apply(lambda x: x.split(",")[0])

# Excluir coluna desnecessária
data.drop('listed_in', axis = 1, inplace = True)

# Mostrar dados
data
```

```python
#Verificando os valores ausentes
data.isnull().sum()
```

```python
# Descartando os registros com valores ausente no dataframe
data = data[~data.duration.isnull()].copy()
data = data[~data.rating.isnull()].copy()
data = data[~data.date_added.isnull()].copy()
```

```python
#Mostrar a tabela dos dados
data
```

```python
#Realizar o gráfico de total por tipo

data.type.value_counts(normalize=True)
data.type.value_counts(normalize=True).plot.pie()
plt.show()
```

```python
#Gráfico para sabermos o total por genero
data.listed_in_novo.value_counts(normalize=True)
data.listed_in_novo.value_counts(normalize=True).plot.barh()
plt.show()
```

```python
#Divisão das datas de lançamentos

data.release_year.describe()
```

```python
#Total de produção por tipo e data de lançamento
plt.scatter(data.release_year,data.type)
plt.show()
```

```python
#Total de genero por data de lançamento
plt.scatter(data.listed_in_novo,data.release_year)
plt.xticks(rotation=90)
plt.show()
```

```python
#Boxplot

sns.boxplot(data.type, data.release_year)
plt.show()
```

# **2) Análise de uma série temporal**

```python
#Série temporal de lançamentos por total de lançamentos

df_grouped = data.groupby("release_year")["title"].count().reset_index()
```

```python
sns.lineplot(data=df_grouped, x="release_year", y="title")
plt.xlabel("Ano de Lançamento")
plt.ylabel("Quantidade de Filmes")
plt.title("Quantidade de Filmes Lançados por Ano")
```

```python
#Série temporal de lançamentos por genero

df_grouped_genero = data.groupby(["release_year", "listed_in_novo"])["title"].count().reset_index()

fig, ax = plt.subplots(figsize=(12, 8))

sns.lineplot(data=df_grouped_genero, x="release_year", y="title", hue="listed_in_novo", ax=ax)
ax.set_xlabel("Ano de Lançamento")
ax.set_ylabel("Quantidade de Filmes")
ax.set_title("Quantidade de Filmes Lançados por Ano e Gênero")
```
