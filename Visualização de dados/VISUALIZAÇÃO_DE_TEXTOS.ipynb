# **Desvendando a Performance dos Pilotos da Fórmula 1: Uma Análise Detalhada dos Resultados**

Análise exploratória de dados

```python
#importar as bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#Ler o conjunto de dados
data= pd.read_csv('https://raw.githubusercontent.com/abreuzzzz/Py_Mackenzie/main/f1_drivers_points%20(1).csv')

#Mostrar os dados
data.head()
```

**Análise dos perfis dos pilotos da temporada 2022 da formula**

```python
# Criar um dicionário com as cores correspondentes a cada país
colors = data['team']

# Criar um gráfico de dispersão cruzando a idade com a nacionalidade

colors = {}
for index, row in data.iterrows():
    if row['team'] == 'RED BULL RACING RBPT':
        colors[row['team']] = 'Blue'
    elif row['team'] == 'FERRARI':
        colors[row['team']] = 'Red'
    elif row['team'] == 'MERCEDES':
        colors[row['team']] = 'Purple'
    elif row['team'] == 'MCLAREN MERCEDES':
        colors[row['team']] = 'Orange'
    elif row['team'] == 'ALPINE RENAULT':
        colors[row['team']] = 'Pink'
    elif row['team'] == 'ALFA ROMEO FERRARI':
        colors[row['team']] = 'Brown'
    elif row['team'] == 'ASTON MARTIN ARAMCO MERCEDES':
        colors[row['team']] = 'Green'
    elif row['team'] == 'HAAS FERRARI':
        colors[row['team']] = 'Yellow'
    elif row['team'] == 'ALPHATAURI RBPT':
        colors[row['team']] = 'Grey'
    elif row['team'] == 'WILLIAMS MERCEDES':
        colors[row['team']] = 'Black'

# Criar um gráfico de dispersão cruzando a idade com a nacionalidade
fig, ax = plt.subplots(figsize=(15, 6))
for index, row in data.iterrows():
    ax.scatter(row['nationality'], row['age'], color=colors[row['team']], s=100)

# Adicionar a legenda das cores
markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colors.values()]
ax.legend(markers, colors.keys(), numpoints=1, bbox_to_anchor=(1.02, 1), loc='upper left')

# Configurar o gráfico
ax.set_xlabel('País', fontsize=12)
ax.set_ylabel('Idade', fontsize=12)
ax.set_title('Idade vs. Nacionalidade dos Pilotos da Fórmula 1', fontsize=14)
ax.grid(True)

# Ajustar o espaço ao redor do gráfico para acomodar a legenda
plt.subplots_adjust(right=0.8)

# Mostrar o gráfico
plt.show()
```

É possível observar que há uma diversidade de nacionalidades representadas, incluindo Holanda, Mônaco, México, Espanha, Grã-Bretanha, França, Austrália, Alemanha, Dinamarca, Canadá, Japão, China e Tailândia.

Além disso, é possível notar que as idades dos pilotos variam de 21 a 40 anos, o que indica uma faixa etária relativamente ampla para esse esporte.

A idade dos pilotos varia entre 21 e 40 anos, com uma média de idade de 27,7 anos. O piloto mais jovem é Mick Schumacher, com apenas 21 anos, enquanto Fernando Alonso é o mais velho, com 40 anos.

A equipe mais jovem é a Haas Ferrari, com uma média de idade de 26,5 anos, enquanto a equipe mais velha é a Alpine Renault, com uma média de idade de 32,5 anos. Isso pode refletir a estratégia de cada equipe em termos de formação de seus pilotos, bem como a sua filosofia em relação à experiência versus juventude.

No geral, a idade dos pilotos é apenas um dos muitos fatores que podem afetar o desempenho na pista, juntamente com a qualidade do carro, a equipe, a habilidade do piloto, a estratégia de corrida e muitos outros fatores.

**Análise do desempenho dos pilotos da temporada 2022 da formula**

```python
# Criar um dicionário com as cores correspondentes a cada país
colors = data['driver']

# Criar um gráfico de dispersão cruzando a idade com a nacionalidade

colors = {}
for index, row in data.iterrows():
    if row['driver'] == 'Max Verstappen':
        colors[row['driver']] = 'Blue'
    elif row['driver'] == 'Charles Leclerc':
        colors[row['driver']] = 'Red'
    elif row['driver'] == 'Sergio Perez':
        colors[row['driver']] = 'Purple'
    elif row['driver'] == 'George Russell':
        colors[row['driver']] = 'Orange'
    elif row['driver'] == 'Carlos Sainz Jr':
        colors[row['driver']] = 'Pink'
    elif row['driver'] == 'Lewis Hamilton':
        colors[row['driver']] = 'Brown'
    elif row['driver'] == 'Lando Norris':
        colors[row['driver']] = 'Green'
    elif row['driver'] == 'Esteban Ocon':
        colors[row['driver']] = 'Yellow'
    elif row['driver'] == 'Fernando Alonso':
        colors[row['driver']] = 'Grey'
    elif row['driver'] == 'Valtteri Bottas':
        colors[row['driver']] = 'Black'
    elif row['driver'] == 'Daniel Ricciardo':
        colors[row['driver']] = 'Magenta'
    elif row['driver'] == 'Sebastian Vettel':
        colors[row['driver']] = 'Cyan'
    elif row['driver'] == 'Kevin Magnussen':
        colors[row['driver']] = 'Navy'
    elif row['driver'] == 'Pierre Gasly':
        colors[row['driver']] = 'Teal'
    elif row['driver'] == 'Lance Stroll':
        colors[row['driver']] = 'Lavender'
    elif row['driver'] == 'Mick Schumacher':
        colors[row['driver']] = 'Olive'
    elif row['driver'] == 'Yuki Tsunoda':
        colors[row['driver']] = 'Maroon'
    elif row['driver'] == 'Zhou Guanyu':
        colors[row['driver']] = 'Turquoise'
    elif row['driver'] == 'Alexander Albon':
        colors[row['driver']] = 'Beige'
    elif row['driver'] == 'Nicholas Latifi':
        colors[row['driver']] = 'Silver'

# Criar um gráfico de dispersão cruzando a idade com a nacionalidade
fig, ax = plt.subplots(figsize=(15, 8))
for index, row in data.iterrows():
    ax.scatter(row['most_overtakes'], row['points'], color=colors[row['driver']], s=100)

# Adicionar a legenda das cores
markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colors.values()]
ax.legend(markers, colors.keys(), numpoints=1, bbox_to_anchor=(1.02, 1), loc='upper left')

# Configurar o gráfico
ax.set_xlabel('Ultrapassagens', fontsize=12)
ax.set_ylabel('Pontuação', fontsize=12)
ax.set_title('Nível de constancia e agressividade dos pilotos da F1', fontsize=14)
ax.grid(True)

# Ajustar o espaço ao redor do gráfico para acomodar a legenda
plt.subplots_adjust(right=0.8)

# Mostrar o gráfico
plt.show()
```

O piloto com mais pontos na temporada é Max Verstappen, com 454 pontos, seguido por Charles Leclerc, com 308 pontos, e Sergio Perez, com 305 pontos. Esses três pilotos lideraram a tabela de pontos ao longo da temporada.

Quanto ao número de ultrapassagens bem-sucedidas, o piloto com mais ultrapassagens é Lewis Hamilton, com 56 ultrapassagens. Fernando Alonso vem em segundo lugar com 54 ultrapassagens, seguido por Zhou Guanyu, com 52 ultrapassagens. Esses pilotos mostraram habilidade em ganhar posições durante as corridas.

Por outro lado, os pilotos com menos pontos e menos ultrapassagens foram Alexander Albon e Nicholas Latifi, com apenas 4 e 2 pontos, respectivamente, e 26 e 12 ultrapassagens bem-sucedidas.

É interessante notar que alguns pilotos tiveram um número significativo de ultrapassagens bem-sucedidas apesar de terem menos pontos do que outros pilotos. Isso pode indicar que esses pilotos tiveram um desempenho abaixo do esperado em classificações e largadas, mas foram capazes de se recuperar durante as corridas.
